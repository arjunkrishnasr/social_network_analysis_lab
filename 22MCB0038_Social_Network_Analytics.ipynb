```python
# Directed Graph 
import networkx as nx 
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt 
%matplotlib inline
```

```python
# Create a Graph
graph = nx.DiGraph()
```

```python
# Add edges
graph.add_edge('A', 'B', weight = 3) 
graph.add_edge('A', 'C', weight = 6) 
graph.add_edge('C', 'A', weight = 6) 
graph.add_edge('B', 'C', weight = 4) 
graph.add_edge('C', 'D', weight = 5) 
graph.add_edge('D', 'E', weight = 2)
```

```python
# Add Attributes
graph.graph['name'] = 'Directed Graph'
graph.graph['type'] = 'directed graph'
```

```python
# Print Attributes
print((graph))
```

```python
nx.draw(graph)
```

```python
# UnDirected Graph
```

```python
# Create a Graph
graph1 = nx.Graph()
```

```python
# Add edges
graph1.add_edge('A', 'B', weight = 3) 
graph1.add_edge('A', 'C', weight = 6) 
graph1.add_edge('C', 'A', weight = 6) 
graph1.add_edge('B', 'C', weight = 4) 
graph1.add_edge('C', 'D', weight = 5) 
graph1.add_edge('D', 'E', weight = 2)

# Add Attributes
graph1.graph['name'] = 'UnDirected Graph'
graph1.graph['type'] = 'undirected graph'
```

```python
nx.draw(graph1)
```

```python
# get adjacency matrix for directed graph
adj_matrix_directed = nx.linalg.graphmatrix.adjacency_matrix(graph).todense()

# get adjacency matrix for undirected graph
adj_matrix_undirected = nx.linalg.graphmatrix.adjacency_matrix(graph1).todense()

# print adjacency matrices
print("Adjacency matrix for directed graph:")
print(adj_matrix_directed)

print("Adjacency matrix for undirected graph:")
print(adj_matrix_undirected)
```
